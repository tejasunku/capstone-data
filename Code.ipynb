```python
import requests
import json
```

```python
base_url = "https://mastodon.social/api/v1/timelines/public"
```

```python
#Test stuff out first
response_json = requests.get(base_url)
response = json.loads(response_json.text)
```

```python
#Check output
response
```

```python
#Grab 100 most recent posts
num_to_grab = 100
max_grab = 40
max_id = None
out = []
while num_to_grab > 0:
    lim = min(num_to_grab, max_grab)
    parms = {'max_id':max_id, 'limit':lim}
    response_json = requests.get(base_url, params = parms)
    response = json.loads(response_json.text)
    out = out + response
    max_id = response[-1]["id"]
    num_to_grab = num_to_grab - max_grab
```

```python
with open("mastodon.json", "w") as out_file:
    json.dump(out,out_file,indent=4)
```
